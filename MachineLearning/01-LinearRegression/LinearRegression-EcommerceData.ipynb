## Linear Regression using Ecommerce Data set

This project handles an Ecommerce custmer data set and contains some features of customer and the yearly amount spent by the customer. Based on this data using a Linear Regression model we approximately predict the yearly amount spent by the customer. 

Data has Customer info, suchas Email, Address, and their color Avatar. Then it also has numerical value columns:

* Avg. Session Length: Average session of in-store style advice sessions.
* Time on App: Average time spent on App in minutes
* Time on Website: Average time spent on Website in minutes
* Length of Membership: How many years the customer has been a member.
* Yearly Amount Spent: Amount spent by the customer yearly

## Linear Regression

```python
#initially import essential libraries for logistic regression and analysis
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df.columns
```

```python
#setting neumerical data as input features and Yearly amount spent as target
X = df[['Avg. Session Length', 'Time on App','Time on Website',
        'Length of Membership']]
Y = df[['Yearly Amount Spent']]
```

```python
X.head()
```

```python
Y.head()
```

```python
## Data train test split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
print(X_train.shape)
print(Y_train.shape)
```

```python
lm = LinearRegression() #creating an instance of linear regression model
lm.fit(X_train,Y_train) #fit the model with train data and its labels
```

```python
#check how good the model is by predicting on the test set
predictions = lm.predict(X_test)
```

```python
#check the relation between real test targets and predictions
plt.scatter(Y_test,predictions) #below scatter plot shows high correlation btwn Y_test and Predictions
plt.xlabel('Target Values')
plt.ylabel('Predicted Values')
```

```python
#Check the model performance using metrics
print('MAE:', metrics.mean_absolute_error(Y_test, predictions))
print('MSE:', metrics.mean_squared_error(Y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, predictions)))
```

```python
metrics.explained_variance_score(Y_test,predictions)#how much variance the model explains
```

```python
#check distribution plot of error(residuals)
#a normally distributed error shows 
#that it was a good selection of linear regression model
sns.distplot((Y_test-predictions),bins=50)
```

```python
#Further evaluation to figure out the importance of each feature analytically
#this is done by checking model coefficients
lm.coef_
```
