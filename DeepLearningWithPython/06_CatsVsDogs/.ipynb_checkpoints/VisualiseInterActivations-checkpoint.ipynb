```python
#importing the model, previously saved 
from keras.models import load_model
model = load_model('cats_and_dogs_small_2.h5')
model.summary() 
```

```python
#loading a single image which was never seen by the network => a test image
img_path = '/home/anish/Documents/Jupyter Notebook/SA-DLwithPy/CatsVsDogs/cats_and_dogs_small/test/cats/cat.1700.jpg'

from keras.preprocessing import image
import numpy as np

#preprocesing the image using image class of keras
img = image.load_img(img_path, target_size=(150, 150))        #resizing
img_tensor = image.img_to_array(img)                          #converting image to tensor (3d - widthxheighxdepth)
img_tensor = np.expand_dims(img_tensor, axis=0)               #adding 1 dimension corresponds to the smaple number 4d tensor is the general input format
img_tensor /= 255.                                            #so that the i/p vlaues lies between 1&0, making it easier for preocessing

print(img_tensor.shape)                                       #shows that the tensor has 1 sample picture of w=150 h=150 d=3
```

```python
#displaying the test image
import matplotlib.pyplot as plt

plt.imshow(img_tensor[0])                   #'0' points to the first element in the img_tensor
plt.show()
```

```python
#instantiating the model by specifying its i/p images and o/p(o/ps of specific layers of convnet ) tesnors
from keras import models

layer_outputs = [layer.output for layer in model.layers[:8]]              #selects first 8 layers of convnet for giving o/p
activation_model = models.Model(inputs=model.input, outputs=layer_outputs)#specifying i/ps and o/ps of the model

activations = activation_model.predict(img_tensor)                        #this will assign i/p as img_tensor for the activation model

first_layer_activation = activations[0]                                   #calling the first element in activation tensor, having o/ps of all 8 layers
print(first_layer_activation.shape)                                       #we can see the o/p of 1st layer in convnet has w=148 h=148 d=32(no of filters)
```

```python
#visualising the 5th channel of the o/p of 1st convnet layer
import matplotlib.pyplot as plt

plt.matshow(first_layer_activation[0, :, :, 4], cmap='viridis')
```

```python
#visualising the 8th channel of the o/p of 1st convnet layer
plt.matshow(first_layer_activation[0, :, :, 7], cmap='viridis')
```

```python
#visualising every channel in every intermidiate activation
layer_names = []                                        #creates a list which contains names of the layers
for layer in model.layers[:8]:
    layer_names.append(layer.name)

#creating a grid for siplying all the images (all filter o/p in each particular layers)
images_per_row = 16

for layer_name, layer_activation in zip(layer_names, activations):
    n_features = layer_activation.shape[-1]
    size = layer_activation.shape[1]
    n_cols = n_features // images_per_row
    display_grid = np.zeros((size * n_cols, images_per_row * size))
    for col in range(n_cols):
        for row in range(images_per_row):
            channel_image = layer_activation[0,:, :, col * images_per_row + row]
            channel_image -= channel_image.mean()
            channel_image /= channel_image.std()
            channel_image *= 64
            channel_image += 128
            channel_image = np.clip(channel_image, 0, 255).astype('uint8')
            display_grid[col * size : (col + 1) * size, 
                         row * size : (row + 1) * size] = channel_image

#plotting all the images
    scale = 1. / size            
    plt.figure(figsize=(scale * display_grid.shape[1],
                    scale * display_grid.shape[0]))
    plt.title(layer_name)
    plt.grid(False)
    plt.imshow(display_grid, aspect='auto', cmap='viridis')

```
